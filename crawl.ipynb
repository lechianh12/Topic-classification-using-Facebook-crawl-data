```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.edge.service import Service
import time
from selenium.common.exceptions import StaleElementReferenceException, NoSuchElementException
from selenium.webdriver.common.keys import Keys
import random

# Đường dẫn tới EdgeDriver
service = Service("E:/NLP project/edgedriver_win64/msedgedriver.exe")
driver = webdriver.Edge(service=service)

# URL của fanpage
fanpage_url = "https://www.facebook.com/banthethaovtv/?locale=vi_VN"
driver.get(fanpage_url)

# Chờ trang tải ban đầu
time.sleep(50)

# Số lần cuộn tối đa để lấy bài viết
max_scrolls = 200

# Tập hợp lưu các bài viết đã xử lý để tránh lặp
save = set()

try:
    for scroll in range(max_scrolls):
        print(f"Cuộn lần {scroll + 1}...")

        # Cuộn xuống cuối trang
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
        time.sleep(random.randint(5, 7))

        # Tìm tất cả các bài viết hiện tại
        try:
            posts = driver.find_elements(By.CSS_SELECTOR, "div[data-ad-preview='message']")
        except Exception as e:
            print(f"Lỗi khi tìm bài viết: {e}")
            continue

        # Xử lý từng bài viết
        for index, post in enumerate(posts, start=1):
            try:
                # Kiểm tra và nhấn nút "Xem thêm" nếu có
                try:
                    see_more_button = post.find_element(By.XPATH, ".//div[contains(text(), 'Xem thêm')]")
                    driver.execute_script("arguments[0].click();", see_more_button)
                    time.sleep(1)  # Chờ để nội dung tải sau khi nhấn
                except NoSuchElementException:
                    pass  # Không có nút "Xem thêm", tiếp tục

                # Kiểm tra nội dung bài viết
                post_text = post.text.strip()
                if post_text == "" or post_text in save:  # Tránh bài viết trống hoặc trùng lặp
                    continue

                # Thêm bài viết vào tập hợp và in ra
                save.add(post_text)
                print(f"\nBài viết {len(save)}:")
                print(post_text)
                print("++++++++++++++++++++++++++++++++++++++++++++++++++++++++++")
            except StaleElementReferenceException:
                print("Bài viết không còn tồn tại, bỏ qua.")
                continue

        # Kiểm tra số bài viết đã lấy
        print(f"Đã tìm thấy {len(save)} bài viết.")
        print("---------------------------------------------------------------")

except Exception as e:
    print(f"Lỗi xảy ra: {e}")

finally:
    driver.quit()
```

```python
import re
import csv

# Hàm loại bỏ emoji và icon
def remove_emojis(text):
    emoji_pattern = re.compile(
        "["
        "\U0001F600-\U0001F64F"  # emoticons
        "\U0001F300-\U0001F5FF"  # symbols & pictographs
        "\U0001F680-\U0001F6FF"  # transport & map symbols
        "\U0001F700-\U0001F77F"  # alchemical symbols
        "\U0001F780-\U0001F7FF"  # Geometric Shapes Extended
        "\U0001F800-\U0001F8FF"  # Supplemental Arrows-C
        "\U0001F900-\U0001F9FF"  # Supplemental Symbols and Pictographs
        "\U0001FA00-\U0001FA6F"  # Chess Symbols
        "\U0001FA70-\U0001FAFF"  # Symbols and Pictographs Extended-A
        "\U00002702-\U000027B0"  # Dingbats
        "\U000024C2-\U0001F251"  # Enclosed characters
        "]+",
        flags=re.UNICODE,
    )
    return emoji_pattern.sub(r"", text)


# Xử lý loại bỏ emoji và icon
result = []
for i in range(len(save)):
    # Chỉ thêm phần tử nếu khác phần tử trước đó
    if i == 0 or save[i] != save[i-1]:
        cleaned_text = remove_emojis(save[i])
        result.append([cleaned_text])

# Xử lý tiếp các ký tự không mong muốn
result = [[item.strip().replace("\n", "") for item in row] for row in result]
result = [[item.strip().replace("-", "") for item in row] for row in result]
result = [[item.strip().replace("\"", " ") for item in row] for row in result]
result = [[item.strip().replace(".", "") for item in row] for row in result]
result = [[item.strip().replace(":(", "") for item in row] for row in result]
result = [[item.strip().replace(":)", "") for item in row] for row in result]


# Loại bỏ khoảng trắng thừa
for i in range(len(result)):
    result[i] = [item.strip() for item in result[i]]

# Ghi vào file CSV
with open("thể thao 3.csv", mode="w", newline="", encoding="utf-8") as file:
    writer = csv.writer(file)
    writer.writerows(result)  # Ghi nhiều dòng

print("File đã được ghi thành công!")
```
