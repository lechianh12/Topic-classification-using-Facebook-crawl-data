```python
from sklearn.feature_extraction.text import TfidfVectorizer
import pandas as pd
from sklearn.model_selection import train_test_split

```

```python
# df = pd.read_csv("/mnt/e/NLP project/final data.csv")
df = pd.read_csv("E:/NLP project/final data.csv")
df.tail()
```

```python
x = df[["Description"]]
y = df[["Label"]]
x
```

```python
y
```

```python
vectorizer = TfidfVectorizer()


x_tfidf = vectorizer.fit_transform(x["Description"])  

```

```python
print(x_tfidf.toarray().shape)
```

```python
print(x_tfidf.toarray())
```

```python
y_class = []
for label in y["Label"]:
    if label == "truyện":
        label = 3
    elif label == "giáo dục":
        label = 0
    elif label == "phim":
        label = 1
    else:
        label = 2
    y_class.append([label])

class_name = ["giáo dục", "phim", "thể thao", "truyện" ]
print(y_class)
```

```python
print(y_class)
```

```python

x_train, x_test, y_train, y_test = train_test_split(x_tfidf, y_class, test_size=0.2, random_state=42, shuffle=True)


print(x_train)
print(y_train)
```

```python
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix




svm_model = SVC(kernel='rbf', probability=True,random_state=42, verbose=True)


svm_model.fit(x_train, y_train)


y_pred = svm_model.predict(x_test)
y_pred2 = svm_model.predict(x_train)

accuracy = accuracy_score(y_test, y_pred)
accuracy2 = accuracy_score(y_train, y_pred2)

print(f'Accuracy : {accuracy * 100:.2f}%')

print(f'Train Accuracy : {accuracy2 * 100:.2f}%')

print("Classification Report :")
print(classification_report(y_test, y_pred))

print("Confusion Matrix :")
print(confusion_matrix(y_test, y_pred))



```

```python
import re
giao_duc_dict = {
    "THPT": "trung học phổ thông",
    "THCS": "trung học cơ sở",
    "CSGT": "cảnh sát giao thông",
    "UBND": "uỷ ban nhân dân",
    "HSG": "học sinh giỏi",
    "GS": "giáo sư",
    "TS": "tiến sĩ",
    "KHTN": "khoa học tự nhiên",
    "TP": "thành phố",
    "BGK": "ban giám khảo",
    "HCV": "huy chương vàng",
    "HSSV": "học sinh sinh viên",
    "SBD": "số báo danh",
    "GDCD": "giáo dục công dân",
    "GD": "giáo dục",
    "NSUT": "nghệ sĩ ưu tú",
    "NTK": "nhà thiết kế",
    "HLV": "huấn luyện viên",
    "CLB": "câu lạc bộ",
    "BXH": "bảng xếp hạng",
    "ĐT": "đội tuyển",
    "GDĐT": "giáo dục đào tạo",
    
}


def remove_emojis(text):
    emoji_pattern = re.compile(
        "[" 
        "\U0001F1E6-\U0001F1FF"  # Regional indicator symbols (flags)
        "\U0001F300-\U0001F5FF"  # Miscellaneous symbols and pictographs
        "\U0001F600-\U0001F64F"  # Emoticons
        "\U0001F680-\U0001F6FF"  # Transport and map symbols
        "\U0001F700-\U0001F77F"  # Alchemical symbols
        "\U0001F780-\U0001F7FF"  # Geometric shapes extended
        "\U0001F800-\U0001F8FF"  # Supplemental arrows-C
        "\U0001F900-\U0001F9FF"  # Supplemental symbols and pictographs
        "\U0001FA00-\U0001FA6F"  # Chess symbols
        "\U0001FA70-\U0001FAFF"  # Symbols and pictographs extended-A
        "\U00002600-\U000026FF"  # Miscellaneous symbols
        "\U00002700-\U000027BF"  # Dingbats
        "\U0001F200-\U0001F2FF"  # Enclosed ideographic supplement
        "\U0001F0A0-\U0001F0FF"  # Playing cards
        "]+", 
        flags=re.UNICODE
    )
    return emoji_pattern.sub(r"", text)


def replace_abbreviations(text, abbreviations):
    text = remove_emojis(text)
    words = text.strip().replace("\n", "") 
    words = text.strip().replace("-", "") 
    words = text.strip().replace("\"", " ")
    words = text.strip().replace(".", "") 
    words = text.strip().replace(":(", "")
    words = text.strip().replace(":)", "") 
    words = text.split()
    replaced_words = [abbreviations.get(word, word) for word in words]
    return ' '.join(replaced_words)
```

```python
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from imblearn.over_sampling import SMOTE

# Cân bằng dữ liệu với SMOTE
smote = SMOTE(random_state=42)
x_train_smote, y_train_smote = smote.fit_resample(x_train, y_train)

# Khởi tạo và huấn luyện MLP
mlp_model = MLPClassifier(hidden_layer_sizes=(100, 50),  # 2 hidden layers với 100 và 50 neurons
                          activation='relu',
                          solver='adam',
                          alpha=0.001,
                          max_iter=300,
                          random_state=42)

mlp_model.fit(x_train_smote, y_train_smote)

# Dự đoán và đánh giá
y_pred = mlp_model.predict(x_test)

accuracy = accuracy_score(y_test, y_pred)

print(f'Accuracy (After SMOTE): {accuracy * 100:.2f}%')

print("Classification Report (After SMOTE):")
print(classification_report(y_test, y_pred))

print("Confusion Matrix (After SMOTE):")
print(confusion_matrix(y_test, y_pred))
```
