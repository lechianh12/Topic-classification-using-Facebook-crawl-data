```python
import csv

# Đọc file CSV và xử lý
with open("phim 3.csv", mode="r", encoding="utf-8") as file:
    reader = csv.reader(file)
    mang = [[row[0]] for row in reader]  # Mỗi phần tử được đặt trong danh sách con

# In ra danh sách
print("Danh sách:", mang)
```

```python
import requests

# URL chứa danh sách từ tiếng Việt
url = "https://raw.githubusercontent.com/stopwords/vietnamese-stopwords/master/vietnamese-stopwords.txt"

# Tải danh sách
response = requests.get(url)
vietnamese_words = response.text.splitlines()

# Chuyển thành tập hợp common words
common_words = set(vietnamese_words)

# Viết hoa tất cả các từ
uppercase_common_words = {word.upper() for word in common_words}

# Xem trước 10 từ
print("Common words viết hoa:", list(uppercase_common_words)[:10])
```

```python
import re


# Tìm từ viết tắt trong văn bản
def find_abbreviations(text):
    #abbreviations = text.lower()
    abbreviations = re.findall(r'\b[A-Z]{2,}\b|\b(?:[A-Z]\.){2,}', text)
    return [abbr for abbr in abbreviations if abbr not in common_words]

# Áp dụng tìm kiếm từ viết tắt
for row in mang:
    for item in row:
        print("Từ viết tắt:", find_abbreviations(item))
```

```python
giao_duc_dict = {
    "THPT": "trung học phổ thông",
    "THCS": "trung học cơ sở",
    "CSGT": "cảnh sát giao thông",
    "UBND": "uỷ ban nhân dân",
    "HSG": "học sinh giỏi",
    "GS": "giáo sư",
    "TS": "tiến sĩ",
    "KHTN": "khoa học tự nhiên",
    "TP": "thành phố",
    "BGK": "ban giám khảo",
    "HCV": "huy chương vàng",
    "HSSV": "học sinh sinh viên",
    "SBD": "số báo danh",
    "GDCD": "giáo dục công dân",
    "GD": "giáo dục",
    "NSUT": "nghệ sĩ ưu tú",
    "NTK": "nhà thiết kế",
    "HLV": "huấn luyện viên",
    "CLB": "câu lạc bộ",
    "BXH": "bảng xếp hạng",
    "ĐT": "đội tuyển",
    "GDĐT": "giáo dục đào tạo",
    
}
#my_list = [["btw"], ["HSG", "r"], ["idk"]]

def replace_abbreviations(text, abbreviations):
    words = text.split()
    replaced_words = [abbreviations.get(word, word) for word in words]
    return ' '.join(replaced_words)

# Xử lý các chuỗi trong danh sách con
result = [[replace_abbreviations(sublist[0], giao_duc_dict)] for sublist in mang]
print(result)



```
