```python
import pandas as pd
df2 = pd.read_csv("./final data.csv")
df2.tail()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Đọc dữ liệu
#df2 = pd.read_csv("./output2.csv")

# Đếm số lần xuất hiện của mỗi nhãn trong cột 'Label'
label_counts = df2['Label'].value_counts()

# Vẽ biểu đồ tròn
plt.figure(figsize=(6, 6))  # Thiết lập kích thước biểu đồ
plt.pie(
    label_counts,  # Số lần xuất hiện của mỗi nhãn
    labels=label_counts.index,  # Tên nhãn
    autopct='%1.1f%%',  # Hiển thị phần trăm
    startangle=90,  # Góc bắt đầu
    colors=['#ff9999', '#66b3ff', '#99ff99', '#ffcc99']  # Màu sắc tùy chỉnh
)

plt.title('Biểu đồ tròn theo số lần xuất hiện của Label')  # Tiêu đề biểu đồ
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Đọc dữ liệu
# df2 = pd.read_csv("./output2.csv")

# Đếm số lần xuất hiện của mỗi nhãn trong cột 'Label'
label_counts = df2['Label'].value_counts()

# Vẽ biểu đồ cột
plt.figure(figsize=(8, 6))  # Thiết lập kích thước biểu đồ
plt.bar(
    label_counts.index,  # Tên nhãn
    label_counts,  # Số lần xuất hiện của mỗi nhãn
    color=['#ff9999', '#66b3ff', '#99ff99', '#ffcc99']  # Màu sắc tùy chỉnh
)

plt.title('Biểu đồ cột theo số lần xuất hiện của Label')  # Tiêu đề biểu đồ
plt.xlabel('Label')  # Nhãn trục x
plt.ylabel('Số lần xuất hiện')  # Nhãn trục y
plt.xticks(rotation=45)  # Xoay các nhãn trên trục x nếu cần thiết
plt.show()
```
